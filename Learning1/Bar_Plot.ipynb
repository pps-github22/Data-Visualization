````python
# importing basic libraries
import numpy as np
import pandas as pd

# Importing matplotlib for data visualization
import matplotlib.pyplot as plt
````

````python
# reading the dataset

# dataset: employee dataset
data = pd.read_csv('Data Sets\employee.csv')
# structure of the dataset
data.head()
````

````python
# we are only going to visualize 1st 500 rows
data = data.loc[:499,:]
print(data.shape)
````

````python
# columns present in the dataset
data.columns
````

Bar Chart

### Univariate Plots

* Where we plot the values of only one variable using a bar plot
* Here, We will take a Variable and we will plot the categories against their values.
* Univarite Analysis is a process of analyzing an attribute present in the data using plots.

````python
# Here we are analyzing the values present in Attrition column of the employee dataset

# we are using value_counts() which return the values related to each of the categories present in the attribute
x = data['Attrition'].value_counts() #returns the values for all the categories present in that attribute

# printing the values
print(x)
````

````python
# using the values in x to plot 

plt.rcParams['figure.figsize'] = (5, 3)
x.plot(kind = 'bar', color ='green')   # with the plot function we plotted these values using a Bar Chart

plt.title('Univariate Analysis on Attrition Column of Employee Dataset', fontsize = 12)
plt.xlabel('Attrition')
plt.ylabel('Values')
plt.show()
````

````python
# analyzing the attribute Department

data['Department'].value_counts().plot(kind = 'bar', color = 'purple')
plt.ylabel('Values')
plt.xlabel('Different Departments')
plt.xticks(rotation = 360)
plt.title('No. of Employees in each Departments', fontsize =12)
plt.show()
````

````python
# analyzing the Education Field

plt.rcParams['figure.figsize'] = (8, 4)
data['EducationField'].value_counts().plot(kind = 'bar', color = 'orange')
plt.ylabel('Values')
plt.xlabel('Different Education Fields')
plt.xticks(rotation = 360)
plt.title('No. of Employees from different Education Fields', fontsize =12)
plt.show()
````

## Bivariate Plots

* Plotting Two Variables in one Plot.
* Bivariate Analysis leaves us with more options to visualize the patterns in data.
* We can compare two different Variables where the First Variable must be categorical in nature and the Second Variable should be continuous variable.
* These Graphs are the most commonly used Bar Graphs.


````python
# we will be using the Bar Function to plot the Bar plots
help(plt.bar)
````

````python
# plot Education vs Monthly Rate

plt.rcParams['figure.figsize'] = (10, 4)
plt.bar(data['EducationField'], data['MonthlyRate'], width = 0.7, align = 'center')
plt.title('Average Salary of Employees from Various Departments')
plt.show()
````

````python
# plot Departments vs Monthly Rate

plt.rcParams['figure.figsize'] = (8, 4)
plt.bar(data['Department'], data['MonthlyRate'], color = 'green', width = 0.55, align = 'center')
plt.title('Average Salary of Employees from Various Education Fields')
plt.show()
````

Multivariate Plots

There are two different types of plots available to visualize Multivariate data using bar Plots
* 1. Grouped Bar Plots
* 2. Stacked Bar Plots

Grouped Bar Plots

````python
# plot Departments vs Monthly Rate

plt.rcParams['figure.figsize'] = (10, 4)
x = data.pivot_table(index = ['Department','EducationField'], values = ['MonthlyRate'], aggfunc = 'mean')
x.plot(kind = 'bar', stacked = False)
plt.xticks(rotation = 90)
plt.show()
````

* Using the Pivot Table Function, The Data is not in good shape for best visualization.
* We can use ```Crosstab() Function``` for better visualization.

We can Plot ```Multi Variate Charts``` for any data using ```crosstab()``` function**


We have two choices, we can plot them as ```Grouped Charts``` and ```Stacked Charts```

````python
# using crosstab funtion, making a cross tabular using multiple variables.
# we have used mean function to aggregate the values of Monthly rate.
# for x axis -> index: Department
# for y axis -> index: Education Field
# values (x X y) -> Monthly Rate, aggreated using mean function

x = pd.crosstab(index = data['Department'],
                columns = data['EducationField'],
                values = data['MonthlyRate'],
                aggfunc = 'mean')
x
````

Stacked Bar Plot

````python
# Stacked Bar Chart
# we will use x, which we have created using crosstab function

# setting the size of the plot
plt.rcParams['figure.figsize'] = (8, 4)

# keeping the Stacked option to **True**
x.plot(kind = 'bar', stacked = True)
plt.ylabel('Salary')
plt.xticks(rotation = 20)
plt.show()
````

Grouped Bar Plot

````python
# Grouped Bar Plot 
# we will be using the variable x which we have created using the cross tab function

# setting the size of the plot
plt.rcParams['figure.figsize'] = (8, 4)

# keeping the stacked option to **false**
x.plot(kind = 'bar', stacked = False)
plt.xticks(rotation = 20)
plt.show()
````
